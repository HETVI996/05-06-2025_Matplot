Matplot:
  Python library for creating static, animated, and interactive 2D plots of data.



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

Basic function

```python
# 1. Line Plot: y = x²
x = np.linspace(0, 5, 1000)
y = x ** 2

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='y = x²', color='blue', linewidth=2)
plt.title("Line Plot: y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 10]

plt.figure(figsize=(6, 4))
plt.bar(labels, values, color='teal')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis='y')
plt.show()
```

```python
# Multiple plot: This is useful when you want to compare different plots side-by-side in the same output.

x = np.linspace(0, 2 * np.pi, 1000)

plt.figure(figsize=(6, 4))
plt.plot(x, np.sin(x), label='sin(x)', color='green')
plt.plot(x, np.cos(x), label='cos(x)', color='red')
plt.title("Multiple Lines: sin(x) and cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Object Oriented/Based Ploting

x = np.linspace(0, 10, 1000)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, color='darkblue', label='sin(x)')
ax.set_title("Object-Based Sine Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
ax.legend()
plt.show()
```

Multiple Figure Creation :

```python
# Data
x = np.linspace(0, 5, 1000)
y1 = x ** 2
y2 = np.sin(x)

# Create figure
fig = plt.figure(figsize=(10, 6), dpi=100)

# Parent plot (main plot)
ax1 = fig.add_axes([0.1, 0.1, 0.85, 0.85])  # Large area
ax1.plot(x, y1, color='blue', linewidth=2)
ax1.set_title("Main Plot: y = x²", fontsize=14)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.grid(True)

# Child/inset plot - repositioned to top-right corner
ax2 = fig.add_axes([0.65, 0.55, 0.25, 0.25])  # Smaller + shifted
ax2.plot(x, y2, color='green', linewidth=1.5)
ax2.set_title("Inset: y = sin(x)", fontsize=10)
ax2.set_xlabel("x", fontsize=9)
ax2.set_ylabel("y", fontsize=9)
ax2.tick_params(labelsize=8)
ax2.grid(True)

plt.show()
```

Functional Programming with Matplotlib:

```python
x = np.linspace(0, 5, 15)
y = x**x
print(
    f"""x: {x}
y: {y}"""
)
```

```python
fig, axes = plt.subplots() #grabbing both figure & axes...

axes.plot(x, y)
```

```python
# For Multiple Figure
fig, axes = plt.subplots(nrows=2, ncols=3)
# plt.tight_layout()

for i in axes:
    for j in i:
        j.plot(x,y)
plt.tight_layout()
```

```python
fig = plt.figure(figsize = (8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y)
```

Legend:

```python
fig, axes = plt.subplots(figsize= (5,5))
axes.plot(x, x**2, label = 'x-squared')
axes.plot(x, x**3, label = 'x-cubed')
axes.legend()
```
