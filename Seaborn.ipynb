   Seaborn is a high-level data visualization library built on top of Matplotlib, designed to make beautiful, informative, and statistically rich plots with minimal code.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Built-in dataset
df = sns.load_dataset('tips')  # Data about restaurant tips
print(df.head())
```

```python
sns.lineplot(data=df, x='total_bill', y='tip')
plt.title("Line Plot of Tip vs Total Bill")
plt.show()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex')
plt.title("Scatter Plot Grouped by Gender")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Boxplot of Bills by Day")
plt.show()
```

```python
sns.histplot(df['total_bill'], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
